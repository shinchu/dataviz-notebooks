```python
# ãƒ•ã‚¡ã‚¤ãƒ«ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ç”¨
import urllib.request
import tarfile
from pathlib import Path

# word2vecç”¨
from gensim.models import KeyedVectors
```

```python
# chiVeã®å­¦ç¿’æ¸ˆã¿ãƒ¢ãƒ‡ãƒ«ã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰
# ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã«æ™‚é–“ãŒã‹ã‹ã‚Šã¾ã™
# https://github.com/WorksApplications/chiVe

url = "https://sudachi.s3-ap-northeast-1.amazonaws.com/chive/chive-1.3-mc5_gensim.tar.gz"
filename = Path(url).name

urllib.request.urlretrieve(url, filename)
with tarfile.open(filename, 'r:gz') as tar:
    tar.extractall()

# .kvãƒ•ã‚¡ã‚¤ãƒ«ã‚’æ¤œç´¢
kv_files = list(Path('.').rglob('*.kv'))

if kv_files:
    model_path = kv_files[0]
    dir_name = str(model_path.parent)
    file_name = model_path.name

    print(f"ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒª: {dir_name}")
    print(f"ãƒ•ã‚¡ã‚¤ãƒ«å: {file_name}")

# ãƒ¢ãƒ‡ãƒ«èª­ã¿è¾¼ã¿
    model = KeyedVectors.load(str(model_path))
    print("ãƒ¢ãƒ‡ãƒ«èª­ã¿è¾¼ã¿å®Œäº†")
```

```python
# ç‰¹å®šã®èªã¨ä¼¼ã¦ã„ã‚‹èªã‚’æŠ½å‡ºã™ã‚‹é–¢æ•°

def explore_word(word):
    print(f"\nã€{word}ã€‘ã«é–¢ã™ã‚‹åˆ†æ")
    print("-" * 50)

    if word in model:
        # é¡ä¼¼å˜èª
        print("ä¼¼ã¦ã„ã‚‹å˜èª:")
        for similar_word, score in model.most_similar(word, topn=10):
            print(f"  {similar_word}: {score:.3f}")
    else:
        print(f"'{word}'ã¯ãƒ¢ãƒ‡ãƒ«ã«å«ã¾ã‚Œã¦ã„ã¾ã›ã‚“")
```

```python
explore_word('æ±äº¬')
```

```python
explore_word('ãƒ­ãƒ³ãƒ‰ãƒ³')
```

```python
# èªåŒå£«ã®è¨ˆç®—
# ç¤¾é•· - ä¼šç¤¾ + å­¦æ ¡ = ?
print("=== ç¤¾é•· - ä¼šç¤¾ + å­¦æ ¡ ===")
result = model.most_similar(
    positive=['ç¤¾é•·', 'å­¦æ ¡'],
    negative=['ä¼šç¤¾'],
    topn=5
)
for word, score in result:
    print(f"{word}: {score:.4f}")
```

```python
# ãƒ­ãƒ³ãƒ‰ãƒ³ - ã‚¤ã‚®ãƒªã‚¹ + ãƒ•ãƒ©ãƒ³ã‚¹ = ?
print("\n=== ãƒ­ãƒ³ãƒ‰ãƒ³ - ã‚¤ã‚®ãƒªã‚¹ + ãƒ•ãƒ©ãƒ³ã‚¹ ===")
result = model.most_similar(
    positive=['ãƒ­ãƒ³ãƒ‰ãƒ³', 'ãƒ•ãƒ©ãƒ³ã‚¹'],
    negative=['ã‚¤ã‚®ãƒªã‚¹'],
    topn=5
)
for word, score in result:
    print(f"{word}: {score:.4f}")
```
